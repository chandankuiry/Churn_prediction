```python
# here i drop the ID column .and I chose feature as [AGE,Plan,LastMonthlyBill,LastpaidAmount,Occupation,IncomeLevel,Gender,SubscriberJoinDate,
#LastSatisfactionSurveyScore]
df1 = df1.drop(["ID"], axis=1)
features = df1.drop(["Churn"], axis=1).columns
```

```python
# I only take train and test data .and i splited train(75%) and test(25%)
train, test = train_test_split(df1, test_size=0.25)
```

```python
#we use RandomForestClassifier .First we initialise our model,
#fit it to our dataset using the fit() method, then simply make our predictions using the predict() method.
clf = RandomForestClassifier(n_estimators=30)
clf.fit(train[features], train["Churn"])

# Make predictions
predictions = clf.predict(test[features])
probs = clf.predict_proba(test[features])
display(predictions)
```

```python
score = clf.score(test[features], test["Churn"])
print("Accuracy: ", score)
```

```python
# now make confusion_matrix and a ROC curve to check  into the quality of our results.
get_ipython().magic('matplotlib inline')
confusion_matrix = pd.DataFrame(
    confusion_matrix(test["Churn"], predictions), 
    columns=["Predicted False", "Predicted True"], 
    index=["Actual False", "Actual True"]
)
display(confusion_matrix)

# Calculate the fpr and tpr for all thresholds of the classification
fpr, tpr, threshold = roc_curve(test["Churn"], probs[:,1])
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
#now I check which is most dependent  feature to predict the churn .
fig = plt.figure(figsize=(20, 18))
ax = fig.add_subplot(111)

f = pd.DataFrame(clf.feature_importances_, columns=["importance"])
f["labels"] = features
f.sort_values("importance", inplace=True, ascending=False)
display(f.head(5))

index = np.arange(len(clf.feature_importances_))
bar_width = 0.5
rects = plt.barh(index , f["importance"], bar_width, alpha=0.4, color='b', label='Main')
plt.yticks(index, f["labels"])
plt.show()
```
